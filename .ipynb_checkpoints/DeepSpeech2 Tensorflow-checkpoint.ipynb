```python
import utils
import config
from  utils import create_data_generator as generator
from tensorflow.keras.activations import relu
import tensorflow as tf
from tensorflow.keras.layers import Input, GRU, Bidirectional,Lambda, TimeDistributed,BatchNormalization,ZeroPadding1D, Conv2D,Conv1D, Dense
from tensorflow.keras.models import Sequential, Model
import keras.backend as K
import speechdeep
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
from jiwer import wer
```

```python
def ctc_lambda_func(args):
    y_pred, labels, input_length, label_length = args
    return K.ctc_batch_cost(labels, y_pred, input_length, label_length)
```

```python
input_data = Input(shape = (None, 161), name = 'inputs')
x = input_data
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center=True, scale = True)(x)
x = ZeroPadding1D(padding=(0, 2451))(x)
x = Conv1D(filters = 256, name = 'conv1', kernel_size = 11, padding='valid', strides = 2, activation = 'relu')(x)
x = ZeroPadding1D(padding=(0, 2451))(x)
x = Conv1D(filters = 256, name = 'conv2', kernel_size = 11, padding='valid', strides = 2, activation = 'relu')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)

x = Bidirectional(GRU(512, name = 'RNN1', return_sequences = True, dropout = 0.1), merge_mode='sum')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = Bidirectional(GRU(512, name = 'RNN2', return_sequences = True, dropout = 0.1), merge_mode='sum')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = Bidirectional(GRU(512, name = 'RNN3', return_sequences = True,  dropout = 0.1), merge_mode='sum')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = Bidirectional(GRU(512, name = 'RNN4', return_sequences = True,  dropout = 0.1), merge_mode='sum')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = Bidirectional(GRU(512, name = 'RNN5', return_sequences = True,  dropout = 0.1), merge_mode='sum')(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = BatchNormalization(axis = -1, momentum = 0.99, epsilon = 1e-3, center = True, scale = True)(x)
x = TimeDistributed(Dense(29, activation='softmax', name='base_model_out'))(x)

base_model = Model(inputs=input_data, outputs=x)

labels = Input(name='labels', shape=(None, ), dtype='int32')
input_lengths = Input(name='input_lengths', shape=[1], dtype='int64')
label_lengths = Input(name='label_lengths', shape=[1], dtype='int64')

loss_out = Lambda(ctc_lambda_func, output_shape=(1,), name='ctc')([x, labels, input_lengths, label_lengths])

final_model = Model(inputs=[input_data, labels, input_lengths, label_lengths], outputs=[loss_out])
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999,epsilon=1e-8, clipnorm=5)
final_model.compile(loss={'ctc': lambda y_true, y_pred: y_pred}, optimizer=optimizer)
```

```python
df_train = pd.read_csv("./train_data/train.csv")
df_val = pd.read_csv("./dev_data/dev.csv")
X_val = []
y_val = []
for i in range(len(df_val)):
    filename, spec_length, labels_length, labels_seq = df_val.iloc[i]
    X_val.append(np.load("./dev_data/" + filename +'.npy'))
    y_val.append([int(i) for i in labels_seq.split(' ')])
X_val = np.array(tf.keras.preprocessing.sequence.pad_sequences(X_val, maxlen = 2451, padding='post'))
y_val = np.array(tf.keras.preprocessing.sequence.pad_sequences(y_val, maxlen = 398, padding='post', value = 27))
```

```python
def GreedyDecoder(output, blank_label=28, collapse_repeated=True):
    arg_maxes = np.argmax(output, 2)
    decodes = []
    for i, args in enumerate(arg_maxes):
        decode = []
        for j, index in enumerate(args):
            if index != blank_label:
                if collapse_repeated and j != 27 and index == args[j -1]:
                    continue
                decode.append(index.item())
        decodes.append(decoder(decode))
    return decodes
```

```python
idx_map = utils.idx_mapping()

def decoder(idx):
    text = ""
    for i in idx:
        if i == 27 or i == 28:
            text += ""
        elif i == 0:
            text += " "
        else:
            text += idx_map[i]
    return text
```

```python
batch_size = 32
```

```python
def evaluate(final_model):
    base_model = Sequential()
    for i in range(15):
        base_model.add(final_model.layers[i])
        
    wer_score = 0
    num_batch = len(X_val)//batch_size
    for i in range(0,num_batch):
        small_X_val = X_val[i*batch_size:(i+1)*batch_size]
        small_y_val = y_val[i*batch_size:(i+1)*batch_size]
        y_pred = base_model(small_X_val)
        preds = GreedyDecoder(y_pred)
        texts = [decoder(y) for y in small_y_val]
        wer_score +=  wer(texts, preds)/num_batch
     
    print(f"Word Error Rate (wer): {wer_score}")
    
    print(f"Show some predictions:")
    random_idx =np.random.randint(batch_size, size=(2))
    for i in random_idx:
        print(f" - True text:  {texts[i]}.")
        print(f" - Prediction: {preds[i]}.")
        print(" -"*25)
```

```python
import time
```

```python
loss_history = []
for i in range(30):
    print("*"*100)
    print("*"*100)
    print(f"Epoch:{i+1}")
    
    loss = 0
    start_time = time.time()
    for j in range(df_train.shape[0]//batch_size):
        if j == 313:
            continue
        print(f"     Ep {i+1} - batch:{j+1}/{df_train.shape[0]//batch_size}")
        
        inputs = []
        labels = []
        input_len = []
        label_len = []
        batch = df_train.iloc[j*batch_size:(j+1)*batch_size]
        for k in range(batch_size):
            filename, spec_length, labels_length, labels_seq = batch.iloc[k]
            inputs.append(np.load("./train_data/" + filename +'.npy'))
            labels.append([int(i) for i in labels_seq.split(' ')])
            input_len.append(int(spec_length))
            label_len.append(int(labels_length))
            
            X = {
                'inputs': np.array(tf.keras.preprocessing.sequence.pad_sequences(inputs, maxlen=2451, padding='post')),
                'labels': np.array(tf.keras.preprocessing.sequence.pad_sequences(labels, maxlen=398, padding='post', value = 27)),
                'input_lengths': np.array(input_len),
                'label_lengths': np.array(label_len)
            }
        y = {"ctc":np.zeros(batch_size)}
        history = final_model.fit(X,y)
        loss += history.history["loss"][0]/(df_train.shape[0]//batch_size)
        inputs = []
        labels = []
        input_len = []
        label_len = []
        loss_history.append(history.history["loss"][0])
    running_time  = time.time() -start_time
    print(f"Loss of Epoch {i+1}: {loss}")
    evaluate(final_model)
    print(f"Running time: {round(running_time/60,2)}m")
```
